```python
%matplotlib notebook

import pandas as pd
import numpy as np
import os
from ml_tps.utils.dataframe_utils import get_test_train_X_y
from ml_tps.utils.regression_utils.linear_regression import LinearRegression
from matplotlib import pyplot as plt

dir_path = os.path.dirname(os.path.realpath("exercise_cerveza.ipynb"))
DEFAULT_FILEPATH = f"{dir_path}/../regression_exercise/data/"
DATA_FILENAME = "cervezas.csv"
OBJECTIVE = "Tiempo"
TRAIN_PCTG = 0.6

dataset = pd.read_csv(DEFAULT_FILEPATH + DATA_FILENAME, sep=";")
X_train, y_train, X_test, y_test = get_test_train_X_y(data=dataset, objective=OBJECTIVE, train_pctg=TRAIN_PCTG)

# Multiple regression using both
multiple_classifier = LinearRegression()
multiple_classifier.fit(X_train, y_train)


sampling_data = pd.DataFrame(np.linspace(min(X_train.iloc[:, 0].values), max(X_train.iloc[:, 0].values), 1000))
sampling_data[1] = pd.DataFrame(np.linspace(min(X_train.iloc[:, 1].values), max(X_train.iloc[:, 1].values), 1000))
    
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.scatter(X_train.iloc[:, 0].values, X_train.iloc[:, 1].values, y_train.values, color="blue")
ax.plot(sampling_data[0], sampling_data[1], multiple_classifier.predict(sampling_data))
plt.show()

```
